```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("ecommerce_final.db")
```

```python
query_1 = """
-- Query 1: Average price, rating, and stock by category
SELECT
    category,
    AVG(price) AS avg_price,
    AVG(rating) AS avg_rating,
    AVG(stock) AS avg_stock
FROM products
GROUP BY category
ORDER BY avg_rating DESC;
"""
pd.read_sql_query(query_1, conn)
```

Query 1: Average Product Metrics by Category

This query groups the products table by category and calculates the average price, average rating, and average stock for each category.

It matters because it provides a basic business summary of how categories differ in terms of pricing, customer satisfaction, and inventory levels.

The general takeaway is that categories are not all alike: some categories are more premium, some are rated more highly, and some hold more inventory than others.


```python
query_2 = """
-- Query 2: Average quantity and discounted total by product_id
SELECT
    product_id,
    AVG(cart_item_quantity) AS avg_quantity,
    AVG(cart_item_discounted_total) AS avg_discounted_total
FROM cart_items
GROUP BY product_id
ORDER BY avg_discounted_total DESC;
"""
pd.read_sql_query(query_2, conn)
```

Query 2: Average Cart Metrics by Product

This query groups the cart_items table by product ID and calculates the average quantity purchased and the average discounted total for each product.

It matters because it helps identify which individual products tend to contribute more value inside shopping carts.

The general takeaway is that some products generate much larger cart contributions than others, either because they are bought in higher quantities or because they have higher per-item value.


```python
query_3 = """
-- Query 3: Join products and cart_items to analyze category-level cart performance
SELECT
    p.category,
    AVG(c.cart_item_quantity) AS avg_quantity,
    AVG(c.cart_item_discounted_total) AS avg_discounted_total
FROM cart_items c
JOIN products p
    ON c.product_id = p.product_id
GROUP BY p.category
ORDER BY avg_discounted_total DESC;
"""
pd.read_sql_query(query_3, conn)
```

Query 3: Category-Level Cart Performance Using a Join

This query joins the cart_items table with the products table on product ID and then calculates average quantity and average discounted cart value by product category.

It matters because category-level cart performance cannot be measured from the cart data alone; the join is needed to connect cart behavior to product attributes.

The general takeaway is that some categories contribute more strongly to cart value than others, which can help an e-commerce analyst prioritize categories for merchandising or promotion.


```python
query_4 = """
-- Query 4: Join products and cart_items to analyze brand-level cart performance
SELECT
    p.brand,
    AVG(c.cart_item_quantity) AS avg_quantity,
    AVG(c.cart_item_discounted_total) AS avg_discounted_total,
    AVG(p.rating) AS avg_rating
FROM cart_items c
JOIN products p
    ON c.product_id = p.product_id
GROUP BY p.brand
ORDER BY avg_discounted_total DESC
LIMIT 15;
"""
pd.read_sql_query(query_4, conn)
```

Query 4: Brand-Level Cart Performance Using a Join

This query joins cart_items and products, groups the results by brand, and calculates average cart quantity, average discounted cart value, and average rating for each brand.

It matters because brands are often a key decision variable in e-commerce analysis, and this query shows which brands tend to generate stronger cart-level commercial performance.

The general takeaway is that higher-spending brands are not always the highest-rated brands, which suggests that brand positioning and pricing may influence cart behavior differently than customer ratings do.


```python
query_5 = """
-- Query 5: Join products and cart_items and group into price buckets
SELECT
    CASE
        WHEN p.price < 25 THEN 'Low'
        WHEN p.price < 100 THEN 'Medium'
        ELSE 'High'
    END AS price_bucket,
    AVG(c.cart_item_quantity) AS avg_quantity,
    AVG(c.cart_item_discounted_total) AS avg_discounted_total
FROM cart_items c
JOIN products p
    ON c.product_id = p.product_id
GROUP BY price_bucket
ORDER BY avg_discounted_total DESC;
"""
pd.read_sql_query(query_5, conn)
```

Query 5: Cart Performance by Price Bucket

This query joins cart_items and products, creates Low, Medium, and High price buckets using a CASE statement, and then calculates average quantity and average discounted cart value for each bucket.

It matters because it helps test whether cart behavior differs systematically by product price level.

The general takeaway is that products at different price tiers contribute to carts in different ways, and this can help analysts understand whether customers tend to spend through premium items or through larger quantities of lower-priced items.


```python
query_6 = """
-- Query 6: Use a window function to rank products by average discounted cart item total
SELECT
    p.title,
    p.category,
    AVG(c.cart_item_discounted_total) AS avg_discounted_total,
    RANK() OVER (ORDER BY AVG(c.cart_item_discounted_total) DESC) AS product_rank
FROM cart_items c
JOIN products p
    ON c.product_id = p.product_id
GROUP BY p.title, p.category;
"""
pd.read_sql_query(query_6, conn)
```

Query 6: Ranking Products by Average Discounted Cart Value

This query joins the two tables, calculates the average discounted cart item total for each product, and then uses a window function to rank products from highest to lowest.

It matters because window functions allow ranking without collapsing the logic into a simple top-N sort, and they are useful for identifying relative product performance.

The general takeaway is that a small number of products may account for a disproportionately large share of cart value.


```python
query_7 = """
-- Query 7: Use a window function to rank categories by average cart item quantity
SELECT
    p.category,
    AVG(c.cart_item_quantity) AS avg_quantity,
    DENSE_RANK() OVER (ORDER BY AVG(c.cart_item_quantity) DESC) AS category_rank
FROM cart_items c
JOIN products p
    ON c.product_id = p.product_id
GROUP BY p.category;
"""
pd.read_sql_query(query_7, conn)
```

Query 7: Ranking Categories by Average Quantity Purchased

This query joins cart_items and products, calculates the average quantity purchased for each category, and uses a window function to assign a dense rank.

It matters because it shows which categories tend to be purchased in larger quantities, which may differ from the categories that generate the highest dollar value.

The general takeaway is that quantity leadership and revenue leadership are not always the same, and both perspectives are useful for e-commerce decision-making.


```python
query_8 = """
-- Query 8: Find products priced above the overall average price
SELECT
    product_id,
    title,
    category,
    price
FROM products
WHERE price > (
    SELECT AVG(price)
    FROM products
)
ORDER BY price DESC;
"""
pd.read_sql_query(query_8, conn)
```

Query 8: Products Priced Above the Overall Average

This query uses a subquery to calculate the overall average product price and then returns only the products priced above that benchmark.

It matters because subqueries are a useful way to compare individual observations against an overall summary metric.

The general takeaway is that above-average priced products can be isolated and studied separately to understand premium positioning in the catalog.


```python
query_9 = """
-- Query 9: Find cart items whose discounted total exceeds the average cart item discounted total
SELECT
    cart_id,
    product_id,
    cart_item_quantity,
    cart_item_discounted_total
FROM cart_items
WHERE cart_item_discounted_total > (
    SELECT AVG(cart_item_discounted_total)
    FROM cart_items
)
ORDER BY cart_item_discounted_total DESC;
"""
pd.read_sql_query(query_9, conn)
```

Query 9: Cart Items Above the Average Discounted Cart Value

This query uses a subquery to calculate the average discounted cart item total across all cart items and then returns only the rows that exceed that value.

It matters because it helps identify high-value cart-item observations and demonstrates another use of subqueries for threshold-based filtering.

The general takeaway is that only a subset of cart items account for above-average value, which may be useful for identifying priority products or customer purchase patterns.


```python
query_10 = """
-- Query 10: Grouped business summary by category
SELECT
    p.category,
    COUNT(*) AS row_count,
    AVG(p.price) AS avg_price,
    AVG(p.rating) AS avg_rating,
    AVG(c.cart_item_quantity) AS avg_quantity,
    AVG(c.cart_item_discounted_total) AS avg_discounted_total
FROM cart_items c
JOIN products p
    ON c.product_id = p.product_id
GROUP BY p.category
ORDER BY avg_discounted_total DESC;
"""
pd.read_sql_query(query_10, conn)
```

Query 10: Category-Level Business Summary

This query joins cart_items and products and creates a broad business summary by category, including row count, average price, average rating, average quantity, and average discounted cart value.

It matters because it combines product-level and cart-level metrics in one place and provides a compact summary for comparing overall category performance.

The general takeaway is that categories differ across several dimensions at once, and decision-makers should consider both customer response and commercial contribution when evaluating category strength.
